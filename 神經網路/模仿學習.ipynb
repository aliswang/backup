```python
import tensorflow as tf
```

```python
#載入資料
(x_train,y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train.astype("float32")/255
x_train = x_train.reshape((-1,784))
y_train = tf.one_hot(y_train.reshape(y_train.shape[0],), depth=10)
```

```python
#訓練老師模組
```

```python
def build_teacher_model():
    inp = tf.keras.Input((784,))
    x = tf.keras.layers.Dense(784*2, kernel_regularizer=tf.keras.regularizers.l1(0.))(inp)
    x = tf.keras.layers.Dense(200)(x)
    x = tf.keras.layers.Dense(10, name="out")(x)
    output = tf.nn.softmax(x)
    return tf.keras.Model(inputs=[inp], outputs=[output])
teacher_model = build_teacher_model()
```

```python
teacher_model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
             loss=tf.losses.MeanSquaredError(),
             metrics=["acc"])
teacher_model.fit(x=x_train, y=y_train, batch_size=1000, validation_split=0.1, epochs=50, verbose=1)
teacher_model.save_weights("test_t.h5")
```

```python
#截取老師模型中途的輸出作為答案(part_output)
```

```python
def build_teachermodel_part():
    inp = tf.keras.Input((784,))
    x = tf.keras.layers.Dense(784*2, kernel_regularizer=tf.keras.regularizers.l1(0.))(inp)
    output = tf.keras.layers.Dense(200)(x)
    return tf.keras.Model(inputs=[inp], outputs=[output])
teacher_model = build_teachermodel_part()
teacher_model.load_weights("test_t.h5", by_name=True)
part_output = teacher_model(x_train, training = False)
```

```python
#訓練學生部分模型(通常會較老師模型小)
```

```python
def build_studentmodel_part():
    inp = tf.keras.Input((784,))
    output = tf.keras.layers.Dense(200, name = "s200")(inp)
    return tf.keras.Model(inputs=[inp], outputs=[output])
student_model = build_studentmodel_part()
```

```python
student_model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
             loss=tf.losses.MeanSquaredError())
student_model.fit(x=x_train, y=part_output, batch_size=1000, epochs=200, verbose=1)
student_model.save_weights("test_s.h5")
```
