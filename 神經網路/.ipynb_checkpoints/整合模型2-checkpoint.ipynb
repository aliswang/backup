```python
import tensorflow as tf
import numpy as np
#參數設定
noise_dim = 3072
dense_dim = 3072
l1_reg = tf.keras.regularizers.l1(0.)
noises=np.random.normal(size=[10000, noise_dim])
#載入資料
(x_train,y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
x_train = x_train.astype("float32")/255
x_train = x_train.reshape((50000,3072))[:10000]
```

```python
def build_model():
    inp = tf.keras.Input((noise_dim,))
    #block 1
    x1 = tf.keras.layers.Dense(dense_dim*2,kernel_regularizer=l1_reg)(inp)
    x1 = tf.keras.layers.BatchNormalization()(x1)
    x1 = tf.keras.layers.Dense(dense_dim)(x1)
    x1 = tf.keras.layers.Dense(dense_dim)(x1)
    x1 = tf.keras.layers.BatchNormalization()(x1)
    xh1 = tf.keras.layers.Dense(dense_dim)(inp)
    xh1 = tf.keras.layers.BatchNormalization()(xh1)
    x1 = x1+xh1
    #block 2
    x2 = tf.keras.layers.Dense(dense_dim*2,kernel_regularizer=l1_reg)(inp)
    x2 = tf.keras.layers.BatchNormalization()(x2)
    x2 = tf.keras.layers.Dense(dense_dim)(x2)
    x2 = tf.keras.layers.Dense(dense_dim)(x2)
    x2 = tf.keras.layers.BatchNormalization()(x2)
    xh2 = tf.keras.layers.Dense(dense_dim)(inp)
    xh2 = tf.keras.layers.BatchNormalization()(xh2)
    x2 = x2+xh2
    #block 3
    x3 = tf.keras.layers.Dense(dense_dim*2,kernel_regularizer=l1_reg)(inp)
    x3 = tf.keras.layers.BatchNormalization()(x3)
    x3 = tf.keras.layers.Dense(dense_dim)(x3)
    x3 = tf.keras.layers.Dense(dense_dim)(x3)
    x3 = tf.keras.layers.BatchNormalization()(x3)
    xh3 = tf.keras.layers.Dense(dense_dim)(inp)
    xh3 = tf.keras.layers.BatchNormalization()(xh3)
    x3 = x3+xh3
    
    #combine
    xv = tf.concat([x1, x2, x3, xin], 1)
    xv = tf.keras.layers.Dense(dense_dim*2,kernel_regularizer=l1_reg)(xv)
    xv = tf.keras.layers.Dense(dense_dim/2)(xv)
    xv = tf.keras.layers.Dense(dense_dim/4)(xv)
    xv = tf.keras.layers.Dense(dense_dim/8)(xv)
    xv1 = tf.keras.layers.Dense(1,activation='sigmoid')(xv)
    xv2 = tf.keras.layers.Dense(1,activation='sigmoid')(xv)
    xv3 = tf.keras.layers.Dense(1,activation='sigmoid')(xv)
    x = 0.3*x1*xv1 + 0.3*x2*xv2 + 0.3*x3*xv3
    output = tf.nn.sigmoid(x)

    return tf.keras.Model(inputs=[inp], outputs=[output])

model = build_model()
noise = tf.random.normal([1, noise_dim])
print(model([noise], training=False))
```

```python
model.summary()
```

```python
lr = 1e-1
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr),\
                 loss=tf.keras.losses.BinaryCrossentropy())
model.fit(x=noises, y=x_train, batch_size=1000, epochs=200, verbose=1)
```

```python
#show
import matplotlib.pyplot as plt
from IPython.display import clear_output
%matplotlib inline

fig=plt.figure(figsize=(8, 8))
columns = 4
rows = 5
for i in range(1, columns*rows +1):
    img = model(noises[i:i+1]).numpy()[0].reshape((32,32,3))
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
    plt.axis('off')
plt.show()
```

```python
def build_model_2():
    inp = tf.keras.Input((noise_dim,))

    x = tf.keras.layers.Dense(dense_dim*6,kernel_regularizer=l1_reg)(inp)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Dense(dense_dim*3)(x)
    x = tf.keras.layers.Dense(dense_dim*3)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    xh = tf.keras.layers.Dense(dense_dim*3)(inp)
    xh = tf.keras.layers.BatchNormalization()(xh)
    x1 = x+xh

    x = tf.keras.layers.Dense(dense_dim*6,kernel_regularizer=l1_reg)(x1)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Dense(dense_dim)(x)
    x = tf.keras.layers.Dense(dense_dim)(x)
    x = tf.keras.layers.BatchNormalization()(x)
    xh = tf.keras.layers.Dense(dense_dim)(x1)
    xh = tf.keras.layers.BatchNormalization()(xh)
    x = x+xh

    output = tf.nn.sigmoid(x)

    return tf.keras.Model(inputs=[inp], outputs=[output])

model = build_model_2()
noise = tf.random.normal([1, noise_dim])
print(model([noise], training=False))
```

```python
def build_model_3():
    inp = tf.keras.Input((noise_dim,))

    x1 = tf.keras.layers.Dense(dense_dim*2,kernel_regularizer=l1_reg)(inp)
    x1 = tf.keras.layers.BatchNormalization()(x1)
    x1 = tf.keras.layers.Dense(dense_dim)(x1)
    x1 = tf.keras.layers.Dense(dense_dim)(x1)
    x1 = tf.keras.layers.BatchNormalization()(x1)
    xh1 = tf.keras.layers.Dense(dense_dim)(inp)
    xh1 = tf.keras.layers.BatchNormalization()(xh1)
    x1 = x1+xh1

    x2 = tf.keras.layers.Dense(dense_dim*2,kernel_regularizer=l1_reg)(x1)
    x2 = tf.keras.layers.BatchNormalization()(x2)
    x2 = tf.keras.layers.Dense(dense_dim)(x2)
    x2 = tf.keras.layers.Dense(dense_dim)(x2)
    x2 = tf.keras.layers.BatchNormalization()(x2)
    xh2 = tf.keras.layers.Dense(dense_dim)(x1)
    xh2 = tf.keras.layers.BatchNormalization()(xh2)
    x2 = x2+xh2

    x3 = tf.keras.layers.Dense(dense_dim*2,kernel_regularizer=l1_reg)(x2)
    x3 = tf.keras.layers.BatchNormalization()(x3)
    x3 = tf.keras.layers.Dense(dense_dim)(x3)
    x3 = tf.keras.layers.Dense(dense_dim)(x3)
    x3 = tf.keras.layers.BatchNormalization()(x3)
    xh3 = tf.keras.layers.Dense(dense_dim)(x2)
    xh3 = tf.keras.layers.BatchNormalization()(xh3)
    x = x3+xh3

    output = tf.nn.sigmoid(x)

    return tf.keras.Model(inputs=[inp], outputs=[output])

model = build_model_3()
noise = tf.random.normal([1, noise_dim])
print(model([noise], training=False))
```
