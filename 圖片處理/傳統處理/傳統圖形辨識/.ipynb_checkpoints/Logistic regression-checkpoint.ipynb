```python
import cv2
import numpy as np
import os
from sklearn import datasets
from sklearn import model_selection
from sklearn import metrics
```

```python
tr=[]
te=[]
bn=[]
eo=[]
ly=[]
for root, dirs, files in os.walk("data/band_neutrophile"):
                                #抓取../band_neutrophile_2/資料夾下的所有檔案
    for name in files:
    #迴圈讀取檔名
        img = cv2.imread(os.path.join(root, name))
        #使用opencv 讀取檔案(os.path.join(root, name)為路徑+檔名)
        img2 = cv2.resize(img, (150, 150))
        #大小縮放統一成150*150
        u = img2.reshape(-1, 67500)
        #壓平成一維陣列
        bn.extend(u)
        #全部塞進list的a中

#輸出檢查格式
bn_f = np.array(bn)#list轉為array

#抽取50個
tr = bn_f[:50]
te = bn_f[50:]

for root, dirs, files in os.walk("data/eosinophile"):
    for name in files:
    #迴圈讀取檔名
        img = cv2.imread(os.path.join(root, name))
        #使用opencv 讀取檔案(os.path.join(root, name)為路徑+檔名)
        img2 = cv2.resize(img, (150, 150))
        #大小縮放統一成150*150
        u = img2.reshape(-1, 67500)
        #壓平成一維陣列
        eo.extend(u)
        #全部塞進list的a中

#輸出檢查格式
eo_f = np.array(eo)#list轉為array

#抽取50個
tr = np.append(tr,eo_f[:50], axis=0)
te = np.append(te,eo_f[50:], axis=0)

for root, dirs, files in os.walk("data/lymphocyte"):
    for name in files:
    #迴圈讀取檔名
        img = cv2.imread(os.path.join(root, name))
        #使用opencv 讀取檔案(os.path.join(root, name)為路徑+檔名)
        img2 = cv2.resize(img, (150, 150))
        #大小縮放統一成150*150
        u = img2.reshape(-1, 67500)
        #壓平成一維陣列
        ly.extend(u)
        #全部塞進list的a中

#輸出檢查格式
ly_f = np.array(ly)#list轉為array

#抽取50個
tr = np.append(tr,ly_f[:50], axis=0)
te = np.append(te,ly_f[50:], axis=0)

#0=bn 1=eo 2=ly
#答案產生
ans = []
a = np.full((50, 1), 0)
b = np.full((50, 1), 1)
c = np.full((50, 1), 2)
ans.extend(a)
ans.extend(b)
ans.extend(c)
ans = np.array(ans)
ans_tr = ans
ans_te = ans

#亂數處理
for i in range(len(tr)):
    c1 = np.random.randint(150, size=1)
    c2 = np.random.randint(150, size=1)
    tr[c1], tr[c2] = tr[c2], tr[c1]
    ans_tr[c1], ans_tr[c2] = ans_tr[c2], ans_tr[c1]

for i in range(len(te)):
    c1 = np.random.randint(150, size=1)
    c2 = np.random.randint(150, size=1)
    te[c1], te[c2] = te[c2], te[c1]
    ans_te[c1], ans_te[c2] = ans_te[c2], ans_te[c1]

#格式處理
tr=tr / 255
te=te / 255
```

```python
k = np.arange(10)

X_train = np.asarray(tr, np.float32)
X_test = np.asarray(te, np.float32)
y_train = np.asarray(ans_tr, np.float32)
y_test = np.asarray(ans_te, np.float32)

lr = cv2.ml.LogisticRegression_create()
lr.setTrainMethod(cv2.ml.LogisticRegression_MINI_BATCH)
lr.setMiniBatchSize(1)
lr.setIterations(100)
lr.train(X_train, cv2.ml.ROW_SAMPLE, y_train)
lr.get_learnt_thetas()
ret, y_pred = lr.predict(X_train)
metrics.accuracy_score(y_train, y_pred)
ret, y_pred = lr.predict(X_test)
metrics.accuracy_score(y_test, y_pred)
```
