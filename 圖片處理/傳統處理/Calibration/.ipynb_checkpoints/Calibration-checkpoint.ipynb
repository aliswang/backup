```python
import cv2
import numpy as np
import glob
  
# 300.001
criteria = (cv2.TERM_CRITERIA_MAX_ITER | cv2.TERM_CRITERIA_EPS, 50, 0.001)
  
# 
objp = np.zeros((9*6,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)  # Z0xy
 
obj_points = []    # 3D
img_points = []    # 2D
images = glob.glob('data/*.jpg')
img = cv2.imread(images[0])
#img = cv2.imread('60790b2ee81c7e1d359f97a058779454.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
size = gray.shape[::-1]
ret,corners = cv2.findChessboardCorners(gray, (9,6), None)
if ret:
    obj_points.append(objp)
    corners2 = cv2.cornerSubPix(gray, corners, (11,11), (-1,-1), criteria)  # 
    if corners2!=[]:
        img_points.append(corners2)
    else:
        img_points.append(corners)
  
    cv2.drawChessboardCorners(img, (9,6), corners, ret)   # OpenCV
    #cv2.imshow('img', img)
    plt.imshow(img)
    #cv2.waitKey(50)

    
print(len(img_points))
cv2.destroyAllWindows()
```

```python
def draw(img, corners, imgpts):
    corner = tuple(corners[0].ravel())
    img = cv2.line(img, corner, tuple(imgpts[0].ravel()), (255,0,0), 5)
    img = cv2.line(img, corner, tuple(imgpts[1].ravel()), (255,0,0), 5)
    img = cv2.line(img, corner, tuple(imgpts[2].ravel()), (255,0,0), 5)
    return img
```

```python
def draw2(img, corners, imgpts):
    imgpts = np.int32(imgpts).reshape(-1,2)
    # draw ground floor in green
    img = cv2.drawContours(img, [imgpts[:4]],-1,(0,255,0),-3)
    # draw pillars in blue color
    for i,j in zip(range(4),range(4,8)):
        img = cv2.line(img, tuple(imgpts[i]), tuple(imgpts[j]),(255),3)
        # draw top layer in red color
        img = cv2.drawContours(img, [imgpts[4:]],-1,(0,0,255),3)
    return img
```

```python
axis = np.float32([[0,0,0], [0,3,0], [3,3,0], [3,0,0],[0,0,-3],[0,3,-3],[3,3,-3],[3,0,-3] ])
```

```python
import cv2
import numpy as np
import glob

# 300.001
criteria = (cv2.TERM_CRITERIA_MAX_ITER | cv2.TERM_CRITERIA_EPS, 30, 0.001)

# 
objp = np.zeros((8 * 6, 3), np.float32)
objp[:, :2] = np.mgrid[0:16:8j, 0:12:6j].T.reshape(-1, 2)  # Z0xy
obj_points = []  # 3D
img_points = []  # 2D
images = glob.glob("60790b2ee81c7e1d359f97a058779454.jpg")
for fname in images:

    img = cv2.imread(fname)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    size = gray.shape[::-1]
    ret, corners = cv2.findChessboardCorners(gray, (8, 6), None)
    print(size)
    if ret:
        obj_points.append(objp)


        corners2 = cv2.cornerSubPix(gray, corners, (5, 5), (-1, -1), criteria)  # 

        if corners2.any:
            img_points.append(corners2)
        else:
            img_points.append(corners)

        cv2.drawChessboardCorners(img, (8, 6), corners, ret)  # OpenCV
        plt.imshow(img)

print(len(img_points))

# 
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(obj_points, img_points, size, None, None)

print("ret:", ret)
print("mtx:\n", mtx ) # 
print("dist:\n", dist)  #    distortion cofficients = (k_1,k_2,p_1,p_2,k_3)
print("rvecs:\n", rvecs)  #   # 
print("tvecs:\n", tvecs)  #   # 
print("------------------------------------------")
tot_error = 0
for i in range(len(obj_points)):
    img_points2, _ = cv2.projectPoints(obj_points[i], rvecs[i], tvecs[i], mtx, dist)
    error = cv2.norm(img_points[i], img_points2, cv2.NORM_L2) / len(img_points2) #
    tot_error += error

mean_error = tot_error / len(obj_points)
print("total error: ", tot_error)  #13.38
print("mean error: ", mean_error)  #1.34
#Iphone6s3024*40321.34
```
