```python
from bs4 import BeautifulSoup
import requests
import pickle
import numpy as np
import csv
from IPython.display import clear_output
```

```python
#抓取資源並寫入檔案
count = np.arange(1,4002,1)#總共4001頁
with open('icd9toicd10_2.csv', 'w', newline='') as csvfile:#新增csv檔
    writer = csv.writer(csvfile)
    for i in count:
        #抓取資源
        url = "http://icdsearch.idv.tw/TRANS/php/contacts.php?Pages="+str(i)#鏈接網頁(_GET['']方式取得)
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")#鏈接網頁
        u = soup.find_all('tr')#抓取含有<tr>元件的內容
        #寫入csv內
        writer.writerows(u)
        #顯示抓取進度
        clear_output(wait=True)
        print("進度: "+str(i)+"/4001")
```

```python
li1=[]
li2=[]
li3=[]
li4=[]
li5=[]

with open('icd9toicd10_2.csv', 'r', newline='',encoding='UTF-8') as csvfile2:
    rows = csv.reader(csvfile2)
    for row in rows:
        value1 = str(row[0]).replace("<td>","").replace("</td>","")
        value2 = str(row[1]).replace("<td>","").replace("</td>","")
        value3 = str(row[2]).replace("<td>","").replace("</td>","")
        value4 = str(row[3]).replace("<td>","").replace("</td>","")
        value5 = str(row[4]).replace("<td>","").replace("</td>","")
        #去除icd9編碼中的小數點
        out = ""
        if(len(value1) > 3):
            for i in range(len(value1)):
                if(i != 3):
                    out += value1[i]
        else:
            out = value1
        #去除icd10編碼中的小數點
        out2 = ""
        if(len(value2) > 3):
            for i in range(len(value2)):
                if(i != 3):
                    out2 += value2[i]
        else:
            out2 = value2
        li1.append(out)
        li2.append(out2)
        li3.append(value3)
        li4.append(value4)
        li5.append(value5)
    #清除csv雜質(標頭)
    li1=list(filter(lambda a: a != 'ICD編碼', li1))
    li2=list(filter(lambda a: a != 'ICD0編碼', li2))
    li3=list(filter(lambda a: a != 'ICD9內容', li3))
    li4=list(filter(lambda a: a != 'ICD10內容', li4))
    li5=list(filter(lambda a: a != '對應情形', li5))
    print(np.array(li1).shape)
    print(np.array(li2).shape)
    print(np.array(li3).shape)
    print(np.array(li4).shape)
    print(np.array(li5).shape)
with open('icd9toicd10_2_new.csv', 'w', newline='') as csvfile:#新增csv檔
    writer = csv.writer(csvfile)
    for i in range(len(li1)):
        u = [[li1[i],li2[i],li3[i],li4[i],li5[i]]]
        writer.writerows(u)
```

```python
#轉換成dict格式
icd = {}
with open('icd9toicd10.csv', 'r', newline='') as csvfile2:
    rows = csv.reader(csvfile2)
    for row in rows:
        #去除csv內雜質(<td> 和 <tr>)
        key = str(row[0]).replace("<td>","").replace("</td>","")
        value = str(row[1]).replace("<td>","").replace("</td>","")
        #去除csv內雜質(key中的小數點)
        out = ""
        if(len(key) > 3):
            for i in range(len(key)):
                if(i != 3):
                    out += key[i]
        else:
            out = key
        #寫入pkl內
        if out not in icd:#沒有key則新增key
            icd[out] = list()
        icd[out].append(value)#將value新增至dict內
    #清除csv雜質(標頭)
    del icd['ICD編碼']
print(icd)
```

```python
#將dict儲存成pkl檔
with open('icd9_icd10.pkl', 'wb') as file:
    pickle.dump(icd, file)
```
