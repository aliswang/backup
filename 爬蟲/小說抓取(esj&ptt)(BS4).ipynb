```python
import requests
from bs4 import BeautifulSoup
import time
import re
from IPython.display import clear_output
```

```python
#ESJ DOWNLOAD
def esj_catch(url):
    """
    輸入: 網址
    輸出: 標題/內文/下一頁網址
    """
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    head = str(soup.find_all('h2')[0]).replace('<h2>','').replace('</h2>','')
    context = str(soup.find_all('div', class_='forum-content mt-3')[0]).replace('<p>','').replace('</p>','').replace('<div class="forum-content mt-3" id="">','').replace('</div>','')
    next_link = re.findall('href="\S*"', str(soup.find_all('a', class_='btn btn-outline-secondary btn-sm btn-next')[0]))[0].replace('"','').replace('href=','')
    return head,context,next_link

head_list = []
context_list = []
file_name = "某大叔的vrmmo活动记"
next_link = "https://www.esjzone.cc/forum/1543764255/22954.html"
num=600
for i in range(num):
    clear_output(wait=True)
    print(str(i) + " /" + str(num - 1))
    try:
        #抓取資料
        head,context, next_link = esj_catch(next_link)
        head_list.append(head)
        context_list.append(context)
        time.sleep(1)
    except:
        print("out_of_range")
        break

        #寫入檔案
with open(file_name + ".txt", "a") as fp:
    for i,j in zip(head_list, context_list):
        fp.writelines(str(i)+"\n")
        fp.writelines(str(j)+"\n")
```
