```python
def set_sql(string):
    #輸入文字的框框
    element = driver.find_elements_by_xpath("//textarea[@autocorrect = 'off']")#輸入文字
    element[0].send_keys(string)
    #Excute按鈕, 查詢
    element = driver.find_elements_by_xpath("//div[@class ='actions']//button[@data-bind = 'click: tryExecuteQuery, visible: $root.canExecute, enable: $root.queryEditorBlank']")
    element[0].click()
```

```python
#download csv 按鈕
def dowload_csv():
    element = driver.find_elements_by_xpath("//a[@id = 'download-csv']")
    element[0].click()
```

```python
#刪除
def delet_sql_context():
    element = driver.find_elements_by_xpath("//textarea[@autocorrect = 'off']")#輸入文字
    actions = ActionChains(driver)
    actions.move_to_element(element[0]).click().perform()
    element[0].send_keys(Keys.CONTROL,"a")
    element[0].send_keys(Keys.DELETE)
```

```python
def check_download():
    erro_2 = 0
    while(os.path.exists('C:\\Users\\nkust1108\\Downloads\\query_result.csv') != True and \
          os.path.exists("C:\\Users\\nkust1108\\Downloads\\尚未確認的 query_result.crdownload") != True):
        try:
            dowload_csv()
            break
        except:
            time.sleep(5)
            erro_2 += 1
            if(erro_2 >= 180):
                erro_2 = 0
                driver.back()
```

```python
def check_exist():
    errotime = 0
    erro_2 = 0
    while(os.path.exists('C:\\Users\\nkust1108\\Downloads\\query_result.csv') != True):
        time.sleep(5)
        errotime += 5
        if(errotime >= 180):
            errotime = 0
            erro_2 += 1
            check_download()
        if(erro_2 >= 3):
            erro_2 = 0
            driver.back()
```

```python
def get_count(string):
    count = ""
    sql_string = "use ccl_cd104183;select COUNT(*) from "+string
    delet_sql_context()
    time.sleep(2)
    set_sql(sql_string)
    time.sleep(2)
    check_download()
    check_exist()
    os.rename(r'C:\Users\nkust1108\Downloads\query_result.csv',r'C:\Users\nkust1108\Downloads\out.csv')
    with open(r'C:\\Users\\nkust1108\\Downloads\\out.csv', 'r', newline='', encoding='UTF-8') as csvfile2:
        i = 0
        rows = csv.reader(csvfile2)
        for row in rows:
            if(i == 1):
                count = row
            else:
                i += 1
    delet_sql_context()
    time.sleep(2)
    os.remove(r'C:\\Users\\nkust1108\\Downloads\\out.csv')
    c=int(count[0])
    return c
```

```python
def download_csv(db_name, batch_num=900000, start_num=0, stop=False, stop_num=0):
    count = get_count(db_name)
    print("get_count " + str(count))
    loop_num = count / batch_num
    print("file_num:" + str(int(loop_num) + 1))
    if(stop == False):
        stop_num = int(loop_num) + 1
        if(stop_num == 0):
            print("erro stop num")
                return None
    for i in range(start_num, stop_num):
        time.sleep(5)
        offset = i * batch_num
        delet_sql_context()
        time.sleep(2)
        if(db_name == "cd2000_2013"):
            str_out = "use ccl_cd104183;select id,id_sex,id_birthday,func_date,acode_icd9_1,acode_icd9_2,acode_icd9_3 from "+db_name+" order by id limit "+str(batch_num)+" offset "+str(offset)
        else:
            str_out = "use ccl_cd104183;select id,id_sex,id_birthday,in_date,icd9cm_code,icd9cm_code_1,icd9cm_code_2,icd9cm_code_3,icd9cm_code_4 from "+db_name+" order by id limit "+str(batch_num)+" offset "+str(offset)          
        print("num_" + str(i + 1) + "\n" + str_out)
        set_sql(str_out)
        time.sleep(2)
        check_download()
        check_exist()
        newname = 'C:\\Users\\nkust1108\\Downloads\\' + db_name + "_part" + str(i) + ".csv"
        os.rename(r'C:\Users\nkust1108\Downloads\query_result.csv',newname)
```

```python
#編排格式並輸出txt檔案
files = []
out_line = ""
for r, d, f in os.walk('C:\\Users\\nkust1108\\Downloads\\'):
    for file in f:
        if('.csv' in file):
            files.append(os.path.join(r, file))
        elif('.CSV' in file):
            files.append(os.path.join(r, file))
for i in files:
    csv_file = []
    print(i)
    with open(i, 'r') as inp :
        for row in csv.reader(inp):
            csv_file += [row]
        pre = csv_file[0][0][:3]
        out_text = "("
        for j in range(len(csv_file)):
            f = csv_file[j][0][:3]
            if(f == pre):
                out_text += '"'+str(csv_file[j][0])+'"'
                out_text += ","
            else:   
                out_text = out_text[:-1]
                out_text += ")"
                #print(out_text)
                out_line += out_text
                out_line += "\n"
                out_text = "("
                out_text += '"'+str(csv_file[j][0])+'"'
                out_text += ","
            pre=csv_file[j][0][:3]
        file_name = str(i).replace(".csv","_text.txt")
        file_name = file_name.replace("_new","")
        file_name = file_name.replace("./","./out/")
        fp = open(file_name, "w")
        fp.write(out_line)
        fp.close()
        out_line = ""
```
