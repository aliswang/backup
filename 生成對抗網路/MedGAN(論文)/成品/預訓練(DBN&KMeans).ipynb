```python
import tensorflow as tf
import h5py
import numpy as np
import time
from RBM import RBM
_data = np.load("VAE_data.npy").astype(np.float32)
_data.shape, _data.dtype
```

```python
dense_dim = _data.shape[-1]
RBM_hidden_sizes = [int(dense_dim*3/4), int(dense_dim*9/16), int(dense_dim*27/64)] 
rbm_list = []
input_size = dense_dim
#For each RBM we want to generate
for i, size in enumerate(RBM_hidden_sizes):
    print('RBM: ', i, ' ', input_size, '->', size)
    rbm_list.append(RBM(input_size, size, epochs=1000, learning_rate=1e-1))
    input_size = size
```

```python
for rbm in rbm_list:
    print('New RBM:')
    rbm.train(_data) 
    _data = rbm.rbm_outpt(_data)
```

# _Kmeans  

```python
import tensorflow as tf
import numpy as np
import time
from IPython.display import clear_output
from sklearn.cluster import KMeans
x_data = np.load('VAE_data.npy')
BUFFER_SIZE = 60000
BATCH_SIZE = 512
x_data.shape
```

```python
kmeans = KMeans(n_clusters=5).fit(x_data)
y_data = kmeans.labels_
y_data = tf.one_hot(y_data.reshape(y_data.shape[0],), depth=5)
```

```python
y_data.shape
```

```python
dense_dim = x_data.shape[-1]
l2_reg = tf.keras.regularizers.l2(0.1)
def b_model():
    inp = tf.keras.Input((dense_dim,))
    x = tf.keras.layers.Dense(1966, name="d1")(inp)
    x = tf.keras.layers.Dense(1474, name="d2")(x)
    x = tf.keras.layers.Dense(1106, name="d3")(x)
    x = tf.keras.layers.Dense(1106*2, kernel_regularizer=l2_reg, name="sae1")(x)
    x = tf.keras.layers.BatchNormalization(name="sae2")(x)
    x = tf.keras.layers.Dense(dense_dim, name="sae3")(x)
    xo = tf.keras.layers.Dense(5, activation='softmax',name="o")(x)

    return tf.keras.Model(inputs=[inp],outputs=[xo])
model=b_model()
model.load_weights("2622_1106.h5", by_name = True)
noise = tf.random.normal([1, dense_dim])
print(model([noise], training=False))
```

```python
lr = 1e-2
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr),\
       loss=tf.keras.losses.CategoricalCrossentropy(),\
       metrics=['accuracy'])
history = model.fit(x=x_data, y=y_data, batch_size=int(x_data.shape[0]/10)+1, epochs=1000, verbose=1)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history["accuracy"], label = "acc")
plt.legend()
plt.show()
plt.plot(history.history["loss"], label = "loss")
plt.legend()
plt.show()
```
