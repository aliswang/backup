```python
def train_gen(epoch):
    #=======================
    import tensorflow as tf
    import numpy as np
    data = np.load('VAE_data.npy')
    noise_dim = 2622
    l1_reg = tf.keras.regularizers.l1(0.)
    #===生成器===
    def build_side_model():
        inp = tf.keras.Input((2622,))
        x = tf.keras.layers.Dense(1966, name="d1", trainable=False)(inp)
        x = tf.keras.layers.Dense(1474, name="d2", trainable=False)(x)
        x = tf.keras.layers.Dense(1106, name="d3", trainable=False)(x)
        x = tf.keras.layers.Dense(1106*2, kernel_regularizer=l1_reg, name="sae1", trainable=False)(x)
        x = tf.keras.layers.BatchNormalization(name="sae2", trainable=False)(x)
        x = tf.keras.layers.Dense(2622, name="sae3", trainable=False)(x)
        xo = tf.keras.layers.Dense(5, activation='softmax', name="o", trainable=False)(x)

        return tf.keras.Model(inputs=[inp], outputs=[xo])
    def build_tree_model():
        inp = tf.keras.layers.Input((2622,))
        x = tf.keras.layers.Dense(2622*2, kernel_regularizer=l1_reg, name="tsae1")(inp)
        x = tf.keras.layers.BatchNormalization(name="tsae2")(x)
        x = tf.keras.layers.Dense(1106, name="hidden")(x)
        x = tf.keras.layers.Dense(1474, name="d3")(x)
        x = tf.keras.layers.Dense(1966, name="d2")(x)
        x = tf.keras.layers.Dense(2622, name="d1")(x)
        o = tf.keras.layers.BatchNormalization(name="o")(x)

        return tf.keras.Model(inputs=[inp], outputs=[o])
    def build_generator():
        sidemodel = build_side_model()
        treemodel1 = build_tree_model()
        treemodel2 = build_tree_model()
        treemodel3 = build_tree_model()
        treemodel4 = build_tree_model()
        treemodel5 = build_tree_model()
        
        inp = tf.keras.Input((2622,))
        x1 = treemodel1(inp)
        x2 = treemodel2(inp)
        x3 = treemodel3(inp)
        x4 = treemodel4(inp)
        x5 = treemodel5(inp)

        xv = sidemodel(inp)
        xv1, xv2, xv3, xv4, xv5 = tf.split(xv, 5, axis=1)
        x = x1*xv1 + x2*xv2 + x3*xv3 + x4*xv4 + x5*xv5

        o = tf.nn.tanh(x)
    
        return tf.keras.Model(inputs=[inp], outputs=[o])
    generator=build_generator()
    #===解碼器===
    def b_decode():
        dense_dim = 5245
        xo = tf.keras.Input((2622,))
        x1 = tf.keras.layers.Dense(dense_dim*4, name="layert11", trainable=False)(xo)
        x1 = tf.keras.layers.Dense(dense_dim, name="layert12", trainable=False)(x1)
        x = tf.keras.layers.Dense(dense_dim, name="layert13", trainable=False)(x1)
        x = tf.keras.layers.LayerNormalization(name="lnt1", trainable=False)(x)
        x = tf.nn.relu(x)

        x1 = tf.keras.layers.Dense(dense_dim*4, name="layert21", trainable=False)(x)
        x1 = tf.keras.layers.Dense(dense_dim, name="layert22", trainable=False)(x1)
        x1 = tf.keras.layers.Dense(dense_dim, name="layert23", trainable=False)(x1)
        x = x1
        x = tf.nn.sigmoid(x)

        x1 = tf.keras.layers.Dense(dense_dim*2, name="layert31", trainable=False)(x)
        x1 = tf.keras.layers.Dense(dense_dim, name="layert32", trainable=False)(x1)
        x2 = tf.keras.layers.Dense(dense_dim*2, name="layert33", trainable=False)(x)
        x2 = tf.keras.layers.Dense(dense_dim, name="layert34", trainable=False)(x2)
        x3 = tf.keras.layers.Dense(dense_dim*2, name="layert35", trainable=False)(x)
        x3 = tf.keras.layers.Dense(dense_dim, name="layert36", trainable=False)(x2)

        xv = tf.concat([x1, x2, x3, x], 1)
        xv = tf.keras.layers.Dense(dense_dim/2, name="layert37", trainable=False)(xv)
        xv = tf.keras.layers.Dense(dense_dim/4, name="layert38", trainable=False)(xv)
        xv = tf.keras.layers.Dense(dense_dim/8, name="layert39", trainable=False)(xv)
        xv1 = tf.keras.layers.Dense(1, activation='sigmoid', name="layertv1", trainable=False)(xv)
        xv2 = tf.keras.layers.Dense(1, activation='sigmoid', name="layertv2", trainable=False)(xv)
        xv3 = tf.keras.layers.Dense(1, activation='sigmoid', name="layertv3", trainable=False)(xv)
        x = 0.33*x1*xv1 + 0.33*x2*xv2 + 0.33*x3*xv3

        o = tf.nn.sigmoid(x)

        return tf.keras.Model(inputs=[xo],outputs=[o])
    decode=b_decode()
    decode.load_weights("VAE_6.h5", by_name=True)
    #===鑑別器===
    def build_part_attention_model():
        inp = tf.keras.layers.Input((5245,))
        x = tf.keras.layers.Dense(2622, name="hidden")(inp)
        x = tf.keras.layers.Dense(1966, name="d1")(x)
        x = tf.keras.layers.Dense(1474, name="d2")(x)
        x = tf.keras.layers.Dense(1106, name="d3")(x)
        x = tf.keras.layers.Dense(1106*2, kernel_regularizer=l1_reg, name="sae1")(x)
        x = tf.keras.layers.BatchNormalization(name="sae2")(x)
        x = tf.keras.layers.Dense(1106, name="sae3")(x)
        o = tf.keras.layers.BatchNormalization(name="o")(x)
        return tf.keras.Model(inputs=[inp], outputs=[o])
    def res_net(inp_dim,out_dim):
        inp = tf.keras.Input((inp_dim,))
        x = tf.keras.layers.Dense(inp_dim*2, kernel_regularizer=l1_reg)(inp)
        x = tf.keras.layers.BatchNormalization()(x)
        x = tf.keras.layers.Dense(out_dim)(x)
        xh = tf.keras.layers.Dense(out_dim)(inp)
        xh = tf.keras.layers.BatchNormalization()(xh)
        o = 0.75*x + 0.25*xh
        return tf.keras.Model(inputs=[inp], outputs=[o])
    def build_discriminator():
        treemodelq = build_part_attention_model()
        treemodelv = build_part_attention_model()
        treemodelk = build_part_attention_model()
        res_net1 = res_net(int(1106), int(1106/2))
        res_net2 = res_net(int(1106/2), int(1106/4))
        res_net3 = res_net(int(1106/4), int(1106/8))

        inpout = tf.keras.Input(shape=(5245,))

        xq = treemodelq(inpout)
        xv = treemodelv(inpout)
        xk = treemodelk(inpout)
        xt = tf.keras.layers.Attention(name="att")([xq, xv, xk])
        x = res_net1(xt)
        x = res_net2(x)
        x = res_net3(x)

        o = tf.keras.layers.Dense(1, activation='sigmoid', name="layero")(x)
        return tf.keras.Model(inputs=inpout, outputs=[o])
    #===訓練===
    if(epoch == 0):
        generator.load_weights("gen_old.h5",by_name = True)
    else:
        generator.load_weights("gen_tmp.h5",by_name = True)
        #載入鑑別器
        discriminator = build_discriminator()
        discriminator.load_weights("dis_tmp.h5")
        #模型訓練設定
        y = np.ones((10723,)) #答案
        def generator_loss(real_score, fake_output):
            fake = decode(fake_output)
            fake_score = discriminator(fake)
            return tf.keras.losses.binary_crossentropy(real_score,fake_score)
        generator.compile(optimizer=tf.keras.optimizers.Adam(1e-4),\
                          loss = generator_loss)
        #開始訓練
        print("training gen")
        for i in range(20):
            noises =  data*0.5 + np.random.normal(size=([data.shape[0], noise_dim]))*0.5
            history = generator.fit(x=[noises], y=[y], batch_size=int(10723/10)+1, epochs=10, verbose=0)
            #記錄
            with open("GenLoss.log","a") as w:
                for j in history.history["loss"]:
                    w.write(str(j) + "\n")
        #回合記錄分隔
        with open("GenLoss.log","a") as w:
            w.write("#########################\n")
            
    #===生成訓練檔案===
    dot_len = 1
    noises = np.repeat(data,dot_len,axis=0)*0.5 + np.random.normal(size=([data.shape[0]*dot_len, 2622]))*0.5
    out_data = np.zeros((data.shape[0]*dot_len, 5245))
    print("making train data")
    for i in range(data.shape[0]*dot_len):
        noise= noises[i].reshape((1, -1))
        gen = generator(noise,training=False).numpy()
        gen = decode(gen,training=False).numpy()[0]
        out_data[i] = gen
    np.save("tmp_gendata.npy",out_data)
    generator.save_weights("gen_tmp.h5")
    print("save done")
```

```python
def train_disc():
    import tensorflow as tf
    import numpy as np
    dot_len = 1
    data_fake = np.load("tmp_gendata.npy")
    data = np.load('origin_data.npy')
    x_data = np.append(data_fake,np.repeat(data, dot_len, axis=0), axis=0)
    y = np.append(np.zeros((data.shape[0]*dot_len,)), np.ones((data_fake.shape[0],)), axis=0)
    #打亂順序
    indices = np.arange(x_data.shape[0])
    x_data = x_data[indices]
    y = y[indices]
    
    l1_reg = tf.keras.regularizers.l1(0.)
    #=======================
    def build_part_attention_model():
        inp = tf.keras.layers.Input((5245,))
        x = tf.keras.layers.Dense(2622, name="hidden")(inp)
        x = tf.keras.layers.Dense(1966, name="d1")(x)
        x = tf.keras.layers.Dense(1474, name="d2")(x)
        x = tf.keras.layers.Dense(1106, name="d3")(x)
        x = tf.keras.layers.Dense(1106*2, kernel_regularizer=l1_reg, name="sae1")(x)
        x = tf.keras.layers.BatchNormalization(name="sae2")(x)
        x = tf.keras.layers.Dense(1106, name="sae3")(x)
        o = tf.keras.layers.BatchNormalization(name="o")(x)
        return tf.keras.Model(inputs=[inp], outputs=[o])
    def res_net(inp_dim, out_dim):
        inp = tf.keras.Input((inp_dim,))
        x = tf.keras.layers.Dense(inp_dim*2, kernel_regularizer=l1_reg)(inp)
        x = tf.keras.layers.BatchNormalization()(x)
        x = tf.keras.layers.Dense(out_dim)(x)
        xh = tf.keras.layers.Dense(out_dim)(inp)
        xh = tf.keras.layers.BatchNormalization()(xh)
        o = 0.75*x + 0.25*xh
        return tf.keras.Model(inputs=[inp], outputs=[o])
    def build_discriminator():
        treemodelq = build_part_attention_model()
        treemodelq.load_weights("2622_1106.h5", by_name = True)
        treemodelv = build_part_attention_model()
        treemodelv.load_weights("2622_1106.h5", by_name = True)
        treemodelk = build_part_attention_model()
        treemodelk.load_weights("2622_1106.h5", by_name = True)
        res_net1 = res_net(int(1106),int(1106/2))
        res_net2 = res_net(int(1106/2), int(1106/4))
        res_net3 = res_net(int(1106/4), int(1106/8))

        inpout = tf.keras.Input(shape=(5245,))

        xq = treemodelq(inpout)
        xv = treemodelv(inpout)
        xk = treemodelk(inpout)
        xt = tf.keras.layers.Attention(name="att")([xq, xv, xk])
        x = res_net1(xt)
        x = res_net2(x)
        x = res_net3(x)

        o = tf.keras.layers.Dense(1, activation='sigmoid', name="layero")(x)
        return tf.keras.Model(inputs=inpout, outputs=[o])
    discriminator = build_discriminator()
    try:
        discriminator.load_weights("dis_tmp.h5")
    except:
        print("cannt load disc model")
    #模型訓練設定
    lr = 1e-4
    discriminator.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr),\
                  loss=tf.keras.losses.BinaryCrossentropy(),\
                  metrics=["acc"])
    #開始訓練
    print("training disc")
    history = discriminator.fit(x=x_data, y=y, batch_size=1000, epochs=50, verbose=0)
    #記錄
    with open("DiscLoss.log","a") as w:
        for i,j in zip(history.history["loss"],history.history["acc"]):
            w.write(str(i)+" "+str(j)+"\n")
        w.write("#########################\n")
        
    discriminator.save_weights("dis_tmp.h5")
    print("save done")
```
