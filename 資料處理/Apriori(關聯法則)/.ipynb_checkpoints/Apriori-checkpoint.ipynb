```python
#資料結構(樹型)
def init_tree(check):
    out = {}
    tmp = {}
    check_list = copy.deepcopy(check)
    check_list_tmp = copy.deepcopy(check_list[0])
    check_list.append(check_list_tmp)
    for i in check_list: 
        while(i != []):
            if(i[0] not in list(tmp.keys())):
                if(len(i) == 1):
                    tmp[i[0]] = dict()
                    tmp[i[0]] = 0
                else:
                    tmp[i[0]] = dict()
            tmp = tmp[i[0]]
            i.remove(i[0])
        tmp = out
    return out

def transor_list_recursive(tmp):
    global _string
    global _key_list
    global _value_list
    c = tmp.keys()
    for i in list(tmp.keys()):
        _string += str(i)+","
        try:
            transor_list_recursive(tmp[i])
        except:
            _key_list.append(_string.split(',')[:-1])
            _value_list.append(tmp[i])
            _string = ""

def dict_tree_tensor_list(dic_i, num=3, cut=20):
    global _string
    global _key_list
    global _value_list
    _string = ""
    _key_list = []
    _value_list = []
    out_dic = {}
    transor_list_recursive(dic_i)
    try:
        tmp = _key_list[0]
        out = [list(map(int, list(_key_list[0])))]
        for i in range(1, len(_key_list)):
            if(len(_key_list[i]) < num + 1):
                set_n = num + 1 - len(_key_list[i])
                _out_part = []
                _out_part.extend(tmp[:set_n])
                _out_part.extend(_key_list[i])
                out.append(list(map(int,_out_part)))
                tmp = _out_part
            else:
                out.append(list(map(int, _key_list[i])))
                tmp = _key_list[i]
        for j in range(len(_value_list)):
            if(_value_list[j] >= cut):
                if(tuple(out[j]) not in out_dic):
                    out_dic[tuple(out[j])] = list()
                    out_dic[tuple(out[j])] = _value_list[j]
                else:
                    out_dic[tuple(out[j])] += _value_list[j]
    except:
        pass
    return out_dic
```

```python
def rev_num(num):
    if(num == 0):
        return 0
    else:
        return num + rev_num(num - 1)

def aprioriGen(Lk, k):
    retList = [] # 滿足條件的頻繁項集
    lenLk = len(Lk)
    for i in range(lenLk):
        for j in range(i + 1, lenLk):
            L1 = list(Lk[i])[: k-2]
            L2 = list(Lk[j])[: k-2]
            L1.sort()
            L2.sort()
            if L1 == L2 :
                retList.append(list(set(Lk[i]) | set(Lk[j])))
    return retList

def del_(dic, cut=20):
    q = list(dic.keys())
    for i in q:
        if(dic[i] < cut):
            del dic[i]
            
def count_inp(check_inp, inp):
    _ = check_inp.keys()
    for i in list(check_inp.keys()):
        if(i in inp):
            try:
                count_inp(check_inp[i], inp)
            except:
                check_inp[i] += 1
                
def count_fir(inp):
    global dic
    for i in inp:
        if((i,) not in dic):
            dic[(i,)]=list()
            dic[(i,)]=1
        else:
            dic[(i,)]+=1            
```

```python
time_total = time.time()
tStart = time.time()
dic = {}
cut = 15000
txt_name = "T18I20N2KD5000K.txt"
with open(txt_name,'r') as a: 
    len_txt = sum(1 for line in a)
with open(txt_name,'r') as a:
    for i in range(len_txt):
        b = list(map(int, a.readline().replace('\n','').split(',')))
        count_fir(b)
    del_(dic, cut)
check_dic = {}
num = 1
tEnd = time.time()
print("round" + str(num) + ": " + str(tEnd - tStart))
while(True):
    init()
    tStart2 = time.time()
    with open(txt_name,'r') as c:
        tStart_c = time.time()
        a = list(dic.keys())
        t = []
        con = False
        for i in a:
            if(len(i) == num):
                t.append(i)
        check=aprioriGen(t, num+1)
        print("check_len" + str(len(check)))
        if(check == []):
            print("done")
            break
        check.sort()
        check2 = init_tree(check)
        tStart3 = time.time()
        tEnd_c = time.time()
        print("list done: " + str(tEnd_c - tStart_c))
        for i in range(len_txt):
            b=list(map(int, c.readline().replace('\n','').split(', ')))
            count_inp(check2, b)
        tEnd3 = time.time()
        print("sreach time:" + str(tEnd3 - tStart3))
        dic.update(dict_tree_tensor_list(check2, num, cut=cut))
        num += 1
        time.asctime(time.localtime(time.time()) )
        tEnd2 = time.time()
        print("round" + str(num)+": " + str(tEnd2 - tStart2))
time_total_end = time.time()
print("totla_time:" + str(time_total_end - time_total))
```
